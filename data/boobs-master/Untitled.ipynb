```python
import os
from os import walk
```

```python
filenames = []
for (dirpath, dirnames, filenames) in walk('./images'):
    filenames.extend(filenames)
    break
print(filenames)
```

```python
for f in filenames:
    if '.jpg' in f:
        print(f)
        case = f.split('_')[0]
        filename = f.split('.')[0]
        with open('./bboxes_yolo/{}.txt'.format(case),'r') as f:
            with open('./labels/{}.txt'.format(filename),'w') as f2:
                f2.write(f.read())
```
